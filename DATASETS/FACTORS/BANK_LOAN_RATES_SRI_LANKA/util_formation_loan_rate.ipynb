```python
import pandas as pd

# Read the existing CSV file
existing_data = pd.read_csv('./bank_loan_rate_changes.csv')  # Replace with the actual file path

# Define lists to store extracted values
day_list = []
month_list = []
year_list = []
loan_rate = []

# Loop through each row
for index, row in existing_data.iterrows():
    date_parts = row[0].split('.')
    day, month, year = int(date_parts[0]), int(date_parts[1]), int(date_parts[2])
    day_list.append(day)
    month_list.append(month)
    year_list.append(year)
    loan_rate.append(row[1])

# Add the extracted values as new columns
new_data = pd.DataFrame()

new_data["Year"] = year_list
new_data["Month"] = month_list
new_data["Loan Rate"]= loan_rate

# Save the updated DataFrame to a CSV file
new_data.to_csv('processed_loan_rates.csv', index=False)
```

```python
import pandas as pd

# Load your existing CSV data into a DataFrame
data = pd.read_csv('./processed_loan_rates.csv')

# Create a new DataFrame to store the results
result = pd.DataFrame(columns=['Year', 'Month', 'Loan Rate'])

# Create a date range from January 2000 to September 2023
date_range = pd.date_range(start='1990-01-01', end='2023-11-01', freq='M')

# Initialize the previous month's rate to None
previous_month_rate = None

# Iterate through each month in the date range
for date in date_range:
    year = date.year
    month = date.month

    # Filter data for the current year and month
    month_data = data[(data['Year'] == year) & (data['Month'] == month)]

    if not month_data.empty:
        # If there are multiple loan rates for the month, calculate the mean
        loan_rate = month_data['Loan Rate'].mean()
    else:
        # If the month is empty, use the previous month's rate
        if previous_month_rate is not None:
            loan_rate = previous_month_rate
        else:
            # If there's no previous month's rate, set a default value
            loan_rate = 0.0  # Change this to a suitable default value

    # Update the previous month's rate for the next iteration
    previous_month_rate = loan_rate

    # Convert year and month to integers and format Loan Rate to 2 decimal points
    result = result.append({'Year': int(year), 'Month': int(month), 'Loan Rate': '{:.2f}'.format(loan_rate)}, ignore_index=True)

# Save the result to a new CSV file
result.to_csv('loan_rates_result.csv', index=False)
```
