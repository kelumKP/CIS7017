```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Replace 'your_file.csv' with the actual path to your CSV file
df = pd.read_csv('../DATASETS/CSE_DATA/SHARE_CHANGE/AAF.N0000.csv')

# Convert the 'Trade Date' column to a datetime format
df['Trade Date'] = pd.to_datetime(df['Trade Date'])
```

```python
# Set the x-axis to 'Trade Date' and y-axis to 'Close (Rs.)'
x = df['Trade Date']
y = df['Close (Rs.)']

# Create a line plot
plt.figure(figsize=(12, 6))  # Optional: Adjust the figure size
plt.plot(x, y, label='Close Price (Rs.)')

# Add labels and a title
plt.xlabel('Trade Date')
plt.ylabel('Close Price (Rs.)')
plt.title('Stock Close Price Over Time')

# Show a legend if needed
plt.legend()

# Display the graph
plt.show()
```
