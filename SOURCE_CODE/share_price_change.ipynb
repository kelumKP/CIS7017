```python
import pandas as pd
import os
```

```python

# Path to the directory containing the CSV files
directory_path = '../DATASETS/CSE_DATA/COMPANY_DATA/DAILY_SHARE_TRADING_STATISTICS_SUMMARY_2006_2022/'

# List all CSV files in the directory
filenames = [os.path.join(directory_path, file) for file in os.listdir(directory_path) if file.endswith('.csv')]

# Initialize an empty DataFrame to store the combined data
combined_data = pd.DataFrame()

# Loop through each file and append it to the combined_data DataFrame
for filename in filenames:
    data = pd.read_csv(filename)
    combined_data = combined_data.append(data, ignore_index=True)

combined_data
```

```python
# Select only "Company Name" and "Symbol" columns
selected_data = combined_data[['Symbol']]

# Remove rows with NaN values
selected_data = selected_data.dropna()

# Remove duplicate rows based on the "Symbol" column
selected_data_no_duplicates = selected_data.drop_duplicates(subset='Symbol')

# Define the file path where you want to save the CSV file
output_file_path = 'output_data.csv'

# Export the DataFrame to a CSV file
selected_data_no_duplicates.to_csv(output_file_path, index=False)
```
