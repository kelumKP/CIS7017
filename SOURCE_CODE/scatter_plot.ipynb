```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset from a CSV file
data = pd.read_csv('./BOGA.N0000.csv')

# Select the columns for the scatter plot
columns_to_plot = [
    'Close (Rs.)',
    'TradeVolume',
    'ShareVolume',
    'Turnover (Rs.)',
    'Loan Rate',
    'Inflation',
    'Buy Rate',
    'Election'
]

# Create a pairplot
sns.pairplot(data[columns_to_plot], diag_kind='kde')
plt.suptitle("Pairplot of Features vs Target for BOGA.N0000", y=1.02)  # Add a title

# Display the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset from a CSV file
data = pd.read_csv('./all_data.csv')

# Select the columns for the scatter plot
columns_to_plot = [
    'Close (Rs.)',
    'TradeVolume',
    'ShareVolume',
    'Turnover (Rs.)',
    'Loan Rate',
    'Inflation',
    'Buy Rate',
    'Election'
]

# Create a pairplot with different colors for each graph
sns.pairplot(data, vars=columns_to_plot, hue='Label', diag_kind='kde', palette='Set1')
plt.suptitle("Pairplot of Features vs Target for 46 Shares", y=1.02)  # Add a title

# Display the plot
plt.show()
```
